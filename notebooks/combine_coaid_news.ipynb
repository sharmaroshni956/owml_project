# Combining CoAID Data

```python
import pandas as pd
from pathlib import Path

base_dir = Path("/Users/roshni/Desktop/DrParmar_Project/owml_project/data/CoAID")   # Base CoAID directory

all_fake, all_real = [], []                                                         # Lists to hold fake and real dataframes

for subfolder in base_dir.iterdir():                                                # Iterates over subfolders in CoAID
    if subfolder.is_dir():
        fake_file = subfolder / "NewsFakeCOVID-19.csv"
        real_file = subfolder / "NewsRealCOVID-19.csv"
        
        if fake_file.exists():                                                      # Checks if fake news file exists
            fake_df = pd.read_csv(fake_file)
            fake_df["label"] = 0                                                    # Labeling fake news as 0
            all_fake.append(fake_df)                                                # Append to fake news list
        
        if real_file.exists():                                                      # Checks if real news file exists  
            real_df = pd.read_csv(real_file)
            real_df["label"] = 1                                                    # Labeling real news as 1   
            all_real.append(real_df)                                                # Append to real news list


```

### Combining All Collected Data

```python
if all_fake or all_real:                                                            # Ensures there is at least some data
    coaid_df = pd.concat(all_fake + all_real, ignore_index=True)                    # Combines all fake and real dataframes
    coaid_df.to_csv(base_dir / "CoAID_News_Combined.csv", index=False)              # Saving combined dataframe to CSV
    print(f"Combined CoAID news data saved: {len(coaid_df)} records total")
else:
    print("No CoAID news CSVs found. Check your folder structure.")
```
